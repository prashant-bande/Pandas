```python
# The .agg() Method
fortune = pd.read_csv('fortune1000.csv', index_col='Rank')
sectors = fortune.groupby('Sector')
fortune.head()
```

```python
sectors.agg({'Revenue': sum, 
             'Profits': [sum, max, 'std'],
            'Employees': 'mean'})
```

```python
sectors.agg(['size', 'sum', 'mean', 'max', 'min'])
```

```python
# Iterating through Groups
df = pd.DataFrame(columns=fortune.columns)
```

```python
df
```

```python
for sectors, data in sectors:
    high_rev = data.nlargest(1, 'Revenue')
    df = df.append(high_rev)
```

```python
df
```

```python
cities = fortune.groupby('Location')
df = pd.DataFrame(columns=fortune.columns)
df
```

```python
for city, data in cities:
    high_rev = data.nlargest(1, 'Revenue')
    df = df.append(high_rev)
```
