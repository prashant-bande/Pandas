```python
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
%matplotlib inline
```

# The .plot() Method


```python
bb = data.DataReader(name="BB", data_source='yahoo', start='2007-07-01', end='2008-12-31')
bb.head()
```

```python
bb['Volume'].plot()
```

```python
bb[['Volume', 'High']].plot()
```

```python
bb[['High', 'Low']].plot()
```

# Bar Graphs


```python
bb.head(10)
```

```python
def rank_perform(stockpr):
    if stockpr <=85:
        return "poor"
    elif stockpr > 85 and stockpr < 110:
        return "Good"
    else:
        return "VeryGood"
```

```python
bb["High"].apply(rank_perform).value_counts().plot(kind="bar")
```

```python
bb["High"].apply(rank_perform).value_counts().plot(kind="barh")
```

# Pie Charts


```python
apple = data.DataReader(name="AAPL", data_source='yahoo', start='2007-07-01', end='2008-12-31')
apple.head()
```

```python
apple['Close'].mean()
```

```python
def rank_perf(sharepr):
    if sharepr >= 21.01289095878601:
        return "Above AVG"
    else:
        return "Below AVG"
```

```python
plt.style.use("ggplot")
apple["Close"].apply(rank_perf).value_counts().plot(kind='pie')
```

# Histograms

```python
def custom_round(stock):
    return int(stock * 10 )
```

```python
apple["Open"].apply(custom_round).plot(kind='hist', bins=15)
```
