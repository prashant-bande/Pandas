```python
# A Review of the .apply() Method on Single Columns
def contomillions(number):
    return str(number) + "Millions!"
```

```python
bond['Box_Office'] = bond['Box_Office'].apply(contomillions)
bond['Budget'] = bond['Budget'].apply(contomillions)
```

```python
columns = ['Box Office', 'Budget', 'Bond Actor Salary']
for col in columns:
    bond[col] = bond[col].apply(contomillions)
```

```python
# The .apply() Method with Row Values
bond = pd.read_csv('jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
```

```python
def goodmovie(row):
    actor = row[1]
    budget = row[4]
    
    if actor == "Pierce Brosnan":
        return "The Best"
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable"
    else:
        return "No Clue"

bond.apply(goodmovie, axis=1)
```
